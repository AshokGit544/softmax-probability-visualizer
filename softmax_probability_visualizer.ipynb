```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define Softmax Function
def softmax(x):
    # Subtract max value for numerical stability
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()

# Step 2: Raw scores (logits) for 3 classes
scores = np.array([3.0, 1.0, 0.2])

# Step 3: Apply Softmax
probabilities = softmax(scores)

# Step 4: Print results
print("Raw scores (logits):", scores)
print("Softmax probabilities:", np.round(probabilities, 4))
print("Sum of probabilities:", np.round(probabilities.sum(), 4))

# Step 5: Show detailed calculation
shifted_scores = scores - np.max(scores)
exp_values = np.exp(shifted_scores)

print("\nStep-by-step understanding")
print("Max score:", np.max(scores))
print("Shifted scores:", np.round(shifted_scores, 4))
print("Exponential values:", np.round(exp_values, 4))
print("Sum of exponential values:", np.round(exp_values.sum(), 4))

# Step 6: Visualize probabilities
classes = ['Apple', 'Banana', 'Orange']

plt.figure(figsize=(8, 5))
plt.bar(classes, probabilities)
plt.title("Softmax Probabilities")
plt.ylabel("Probability (0 to 1)")
plt.ylim(0, 1)
plt.show()
```
